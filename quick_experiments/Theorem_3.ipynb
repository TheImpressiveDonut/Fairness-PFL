```python
import numpy as np
from tqdm.notebook import tqdm
```

```python
def verify_theorem_3() -> None:
    N = np.random.randint(2, 100)
    D = 1
    betas_opt = np.random.random(size=(N, D))
    betas = betas_opt + np.random.random(size=(N, D))
    var = np.random.random(size=(N, 1))
    
    K = betas_opt @ betas.T
    C = betas @ betas.T
    V = np.eye(N) * var
    W = K @ np.linalg.inv(C + V)
    
    W_element_wise = np.empty(shape=(N, N))
    E = 1 + (betas ** 2 / var).sum()
    
    for i in range(N):
        for j in range(N):
            W_element_wise[i, j] = (betas_opt[i] * betas[j]).item() / var[j].item()
            
    W_element_wise /= E   
    assert np.allclose(W, W_element_wise)
```

```python
verify_theorem_3()
```

```python
for _ in tqdm(range(10000)):
    verify_theorem_3()
```
